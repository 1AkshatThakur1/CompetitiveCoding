#Find the K-Beauty of a Number

You are given an integer number and a length k. The k-beauty of a number is defined as the number of substrings of the number (when represented as a string) that meet the following conditions:

The substring has a length of k.
The substring, when converted to an integer, is a divisor of the original number.
Write a program to calculate the k-beauty of a given number.

Input:

The input consists of two integers separated by spaces: the number (num) and the length of substrings (k).
1 <= num <= 10^9 (the number to be processed)
1 <= k <= length of num as a string (taking num as a string)
Output:

The output should be an integer representing the k-beauty of the number.
Note:

Leading zeros are allowed in the substrings.
Zero (0) is not considered a divisor of any number.
Input and Output Format:

Refer sample input and output for formatting specifications.

[All text in bold corresponds to input and the rest corresponds to output]

Sample Input and Output:

Enter the number:

240
Enter the length of substrings (k):

2
The k-beauty of the number is: 2

Explanation:
The substrings of length 2 are "24" and "40" from "240".
- 24 is a divisor of 240.
- 40 is a divisor of 240.
Therefore, the k-beauty is 2.

```python
def k_beauty_of_number(num, k):
    num_str = str(num)
    n = len(num_str)
    k_beauty_count = 0

    for i in range(n - k + 1):
        substring = num_str[i:i + k]
        substring_int = int(substring)

        if substring_int != 0 and num % substring_int == 0:
            k_beauty_count += 1

    return k_beauty_count

# Main function to handle input and output
def main():
    # Input number and k
    num = int(input("Enter the number:\n"))
    k = int(input("Enter the length of substrings (k):\n"))

    # Calculate the k-beauty of the number
    result = k_beauty_of_number(num, k)

    # Output the result
    print(f"The k-beauty of the number is: {result}")

if __name__ == "__main__":
    main()
```
