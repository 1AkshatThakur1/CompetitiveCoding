#Insert Greatest Common Divisors in Linked List

In the realm of linked lists, a magical task awaits your coding expertise. You are assigned the mission of enhancing a linked list by inserting nodes with values representing the Greatest Common Divisor (GCD) between adjacent nodes. Your quest is to create a program that brings this mystical transformation to life.

Define a structure

ListNode {
           int val;
           ListNode next;
}

Include functions:

ListNode insert(ListNode head) --- This function takes the head of a linked list as input and inserts nodes with values representing the Greatest Common Divisor (GCD) between adjacent nodes. It returns the modified linked list after insertion.
int gcd(int x, int y) --- This function calculates and returns the Greatest Common Divisor (GCD) of two given integers, x and y.
void printList(ListNode head) --- This function prints the values of a linked list starting from the provided head node.
Constraints:
The number of nodes in the list is in the range [1, 5000].
1 <= Node.val <= 100
Input and Output Format:
Refer sample input and output for formatting specifications.

[All text in bold corresponds to input and the rest corresponds to output.]

Sample Input and Output :

Enter the number of nodes in the linked list:
4
Enter the value for node 1:
18
Enter the value for node 2:
6
Enter the value for node 3:
10
Enter the value for node 4:
3

Linked List After Inserting GCD Nodes:
18 6 6 2 10 1 3

```python
import math

# Definition for singly-linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def gcd(x, y):
    """ Function to compute the Greatest Common Divisor (GCD) of x and y """
    return math.gcd(x, y)

def insert(head):
    """ Function to insert nodes representing GCD between adjacent nodes """
    if not head or not head.next:
        return head

    current = head
    while current and current.next:
        gcd_value = gcd(current.val, current.next.val)
        new_node = ListNode(gcd_value)
        new_node.next = current.next
        current.next = new_node
        current = current.next.next

    return head

def printList(head):
    """ Function to print the values of a linked list """
    current = head
    while current:
        print(current.val, end=" ")
        current = current.next
    print()

# Function to create a linked list from a list of values
def createList(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for val in values[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Main function to test the implementation
if __name__ == "__main__":
    # Example usage
    n = int(input("Enter the number of nodes in the linked list: "))
    values = []
    for i in range(1, n + 1):
        value = int(input(f"Enter the value for node {i}: "))
        values.append(value)

    head = createList(values)
    print("\nLinked List After Inserting GCD Nodes:")
    head = insert(head)
    printList(head)
```
