#Remove Zero Sum Consecutive Nodes

In the mystical realm of Linked Lists, there existed a magical list known as the Summara. Each node in the Summara held a unique integer value, creating a tapestry of numbers woven together. Legend spoke of a mystical ritual that would traverse this list and eliminate consecutive sequences of nodes whose values summed to zero. The chosen adventurer, you, must embark on a quest to invoke this enchantment and unveil the final form of the linked list.



Define a structure

ListNode {
               int val;
              ListNode next;
}



Include functions:

ListNode createLinkedList() --- This function is responsible for creating a linked list by taking user inputs for the values of each node.

ListNode removeZeroSumConsecutiveNodes(ListNode head) --- This function is responsible for creating a linked list by taking user inputs for the values of each node.

void printList(ListNode head) --- This function prints the values of the linked list

Constraints:
The Summara, the linked list in question, will contain between 1 and 1000 nodes.
Each node within the Summara holds an integer value ranging from -1000 to 1000.


Input and Output Format:
Refer sample input and output for formatting specifications.

[All text in bold corresponds to input and the rest corresponds to output.]

Sample Input and Output :

Enter the number of nodes:
5
Enter the values for each node:
1
2
-3
3
1

Original List
1 2 -3 3 1

List after removing zero sum consecutive nodes:
3 1

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def createLinkedList():
    print("Enter the number of nodes:")
    n = int(input().strip())
    if n == 0:
        return None

    print("Enter the values for each node (space-separated):")
    values = list(map(int, input().strip().split()))

    if len(values) != n:
        raise ValueError("Number of values does not match the specified number of nodes.")

    head = ListNode(values[0])
    current = head
    for val in values[1:]:
        new_node = ListNode(val)
        current.next = new_node
        current = new_node

    return head

def printList(head):
    if not head:
        print("List is empty.")
        return

    current = head
    result = []
    while current:
        result.append(str(current.val))
        current = current.next
    print(" ".join(result))

def removeZeroSumConsecutiveNodes(head):
    # Dummy node to handle cases where the head itself needs to be removed
    dummy = ListNode(0)
    dummy.next = head

    prefix_sum = 0
    prefix_map = {}
    current = dummy

    while current:
        prefix_sum += current.val

        if prefix_sum in prefix_map:
            # Remove nodes between prefix_map[prefix_sum] and current
            prev = prefix_map[prefix_sum]
            remove_sum = prefix_sum
            while prev != current:
                remove_sum += prev.next.val
                del prefix_map[remove_sum]
                prev = prev.next
            prev.next = current.next
        else:
            prefix_map[prefix_sum] = current

        current = current.next

    return dummy.next

# Main function to test the implementation
if __name__ == "__main__":
    head = createLinkedList()

    if not head:
        print("Original List\nList is empty.")
    else:
        print("Original List")
        printList(head)

        head = removeZeroSumConsecutiveNodes(head)

        print("\nList after removing zero sum consecutive nodes:")
        printList(head)
```
