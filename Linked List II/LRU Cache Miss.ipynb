#LRU Cache Miss

The Least Recently Used(LRU) cache algorithm exists the element from the cache(when it's full) that was least recently used. After an element is requested from the cache, it should be added to the cache(if not already there) and considered the most recently used element in the cache. Initially, the cache is empty. The input to the function LruCountMiss shall consist of an integer max_cache_size, an array pages and its length. The function should return an integer for the number of cache misses using the LRU cache algorithm.
Assume that the array pages always has pages numbered from 1 to 50.

Include functions:
int lruCountMiss(int , int *, int) -- to count the missing pages from the array.

Input Format:
First line represents integer which corresponds to the Cache size.
Second line represents integer which corresponds to 'n' i.e Array size.
Third line represents 'n' integer values which corresponds to Array elements.

Output Format:
Output consists of a integer which corresponds to number of cache misses.

Note: All text in bold corresponds to input and the rest corresponds to output.

Sample Input 1:
3
16
7 0 1 2 0 3 0 4 2 3 0 3 2 1 2 0
Sample Output 1:
Miss:11

Sample Input 2:
2
9
2 3 1 3 2 1 4 3 2
Sample Output 2:
Miss:8


```python
def lruCountMiss(max_cache_size, pages, n):
    cache = []
    cache_set = set()
    miss_count = 0

    for page in pages:
        if page in cache_set:
            # Page is already in cache (cache hit), update its position in LRU order
            cache.remove(page)
            cache.append(page)
        else:
            # Page is not in cache (cache miss)
            miss_count += 1
            if len(cache) >= max_cache_size:
                # Remove the least recently used page from cache
                evicted_page = cache.pop(0)
                cache_set.remove(evicted_page)
            # Add current page to cache (most recently used)
            cache.append(page)
            cache_set.add(page)

    return miss_count

# Sample Input and Output handling
if __name__ == "__main__":
    max_cache_size = int(input())
    n = int(input())
    pages = list(map(int, input().split()))

    miss_count = lruCountMiss(max_cache_size, pages, n)
    print(f"Miss:{miss_count}")
```
