#Find the  K-sum of an Array

You are given an integer array nums and a positive integer k . You can choose any subsequence of the array and sum all of its elements together.

Define the K-Sum of the array as the kth largest subsequence sum that can be obtained (not necessarily distinct).

Return the K-Sum of the array.

A subsequence is an array that can be derived from another array by deleting some or no elements without changing the order of the remaining elements.

Note that the empty subsequence is considered to have a sum of 0.

Input Format:

The first line contains an integer n, denoting the size of the array nums.
The second line contains n space-separated integers, representing the elements of the array nums.
The third line contains an integer k, representing the kth largest subsequence sum to find.
Output Format:

Output a single integer, representing the K-Sum of the array.
Constraints:

1 <= n <= 2 * 10^4
-1000 <= nums[i] <= 1000
1 <= k <= n
Input and Output Format:

Refer to the sample input and output for formatting specifications.
Sample Input and Output:

Enter the size of the array

3

Enter the array elements separated by space

2 4 -2
Enter the value of k

5
Output: 2
Explanation:

nums = [2, 4, -2]
k = 5
The task is to find the 5th largest subsequence sum that can be obtained from the array nums.

The possible subsequences and sum of the given array:   [2]=2, [4]=4, [-2]=-2, [2, 4]=6, [2, -2]=0, [4, -2]=2, [2, 4, -2]=4.

By sorting these sums in descending order, we get,    [6, 4, 4, 2, 2, 0, -2].

The 5th largest subsequence sum from this list is 2.

```python
def kSum(nums, k):
    # Generate all possible subsequence sums
    sums = set()
    for i in range(len(nums)):
        for j in range(i, len(nums)):
            sums.add(sum(nums[i:j+1]))

    # Sort the sums in descending order
    sorted_sums = sorted(sums, reverse=True)

    # Check if k is within the range of sorted_sums
    if k <= len(sorted_sums):
        # Return the Kth largest sum
        return sorted_sums[k-1]
    else:
        return "Invalid value of k"

def main():
    print("Enter the size of the array:")
    n = int(input())
    print("Enter the array elements separated by space:")
    nums = list(map(int, input().split()))
    print("Enter the value of k:")
    k = int(input())

    result = kSum(nums, k)
    print("Output:", result)

if __name__ == "__main__":
    main()
```
