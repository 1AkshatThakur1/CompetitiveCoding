#Split array


Write a program to Split an array based on the mid element.



Procedure:

Step 1: Declare an integer for size of an array i.e n, one integer array and an integer to store the  mid_index.

Step 2: Print a statement as "Enter the size of array".

Step 3: Read the size of the array.

Step 4: Print the statement as "Enter the elements of array".

Step 5: Read the elements of the array in a for loop.

Step 6: Find the mid index of the array and store it in mid_index.

Step 7: Declare 2 new arrays to store the split arrays.

Step 8: Traverse the original array from index 0 to mid_index-1 and store the array elements in one of the new arrays.

Step 9: Traverse the original array from index mid_index to size-1 and store the array elements in another new array.

Step 10: Display elements of both the arrays after splitting with proper message.



Hint:

mid will be the half of the total size of array and always an integer value.
Example: If size of array is 5, mid will be 5/2 = 2
The new sizes of both arrays will be less than the original array.
If size of original array is odd, the sizes will be half and half+1.
If size of original array is even, both the sizes will be half.



Input and Output Format:
Refer sample input and output for formatting specifications.
[All text in bold corresponds to input and the rest corresponds to output.]

Sample Input and Output :

Enter the size of array

5

Enter the elements of array

5 8 10 14 21

Elements after splitting
5 8
10 14 21

```python
def main():
    print("Enter the size of array:")
    n = int(input())
    print("Enter the elements of array:")
    arr = list(map(int, input().split()))

    mid_index = n // 2
    first_half = arr[:mid_index]
    second_half = arr[mid_index:]

    print("Elements after splitting")
    print(" ".join(map(str, first_half)))
    print(" ".join(map(str, second_half)))

if __name__ == "__main__":
    main()
```
