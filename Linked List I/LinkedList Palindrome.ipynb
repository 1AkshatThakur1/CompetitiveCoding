#LinkedList Palindrome
Function to check if a singly linked list is palindrome
Given a singly linked list of characters, write a function that returns true if the given list is a palindrome, else false.

METHOD1(UseaStack):
A simple solution is to use a stack of list nodes. This mainly involves three steps.
1) Traverse the given list from head to tail and push every visited node to stack.
2) Traverse the list again. For every visited node, pop a node from the stack and compare data of popped node with currently visited node.
3) If all nodes matched, then return true, else false.
Time complexity of above method is O(n), but it requires O(n) extra space.



Input Format:

First Line Consists of the input string str.

Constraints:

String(str) length should be less than 100 characters.
Str is CASE SENSITIVE.


Output Format:

Display whether the string(str) is palindrome or Not.



Sample Input 1:

Wow

Sample Output 1:

W NULL
Is Palindrome

o W NULL
Not Palindrome

w o W NULL
Not Palindrome

Sample Input 2:

ladder

Sample Output 2:

l NULL
Is Palindrome

a l NULL
Not Palindrome

d a l NULL
Not Palindrome

d d a l NULL
Not Palindrome

e d d a l NULL
Not Palindrome

r e d d a l NULL
Not Palindrome

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.stack = []

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def is_palindrome(self):
        slow = self.head
        fast = self.head

        # Traverse the list with fast pointer to find the middle
        while fast and fast.next:
            self.stack.append(slow.data)
            slow = slow.next
            fast = fast.next.next

        # If the length of the list is odd, move slow one step forward
        if fast:
            slow = slow.next

        # Compare the remaining elements with stack
        while slow:
            if slow.data != self.stack.pop():
                return False
            slow = slow.next

        return True

def main():
    # Input string
    input_str = input().strip()

    # Create a linked list from the input string
    ll = LinkedList()
    for char in input_str:
        ll.push(char)

    # Check if the linked list (string) is a palindrome
    if ll.is_palindrome():
        print("Is Palindrome")
    else:
        print("Not Palindrome")

if __name__ == "__main__":
    main()
```
