#Reverse Linked List

A minimal singly linked list implementation is given and also a stack is provided, that can store addresses of linked list nodes. The program is expected to create a reversed linked list from the given list with no additional nodes used/created.
Fill in the function  reverseList() to create a reversed list with help of given stack.The function must return the starting address of the reversed linked list. You are not expected to change any other parts of the code.

Note: The text in bold indicates input.
Sample input & output:
Sample 1:
How many numbers to insert into list:5
Enter 5 numbers:1 2 3 4 5
The list is:1 2 3 4 5
The reverse list is:5 4 3 2 1

Sample 2:
How many numbers to insert into list:2
Enter 2 numbers:3 4
The list is:3 4
The reverse list is:4 3

```python
# Node class to define each node of the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to reverse the linked list using a stack
def reverseList(head):
    if not head or not head.next:
        return head

    stack = []
    current = head

    # Push all nodes onto the stack
    while current:
        stack.append(current)
        current = current.next

    # Create the reversed list by popping from stack
    new_head = stack.pop()
    current = new_head

    while stack:
        current.next = stack.pop()
        current = current.next

    current.next = None  # Mark the end of the reversed list

    return new_head

# Function to print the linked list
def printList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Main function to run the program
def main():
    # Input number of nodes and their values
    n = int(input("How many numbers to insert into list:"))
    if n <= 0:
        print("Invalid input. Number of nodes must be greater than zero.")
        return

    print(f"Enter {n} numbers:", end=" ")
    values = list(map(int, input().strip().split()))

    # Create the original linked list
    if len(values) != n:
        print("Invalid input. Number of values doesn't match.")
        return

    head = Node(values[0])
    current = head
    for value in values[1:]:
        new_node = Node(value)
        current.next = new_node
        current = new_node

    # Print the original list
    print("The list is:", end=" ")
    printList(head)

    # Reverse the list
    new_head = reverseList(head)

    # Print the reversed list
    print("The reverse list is:", end=" ")
    printList(new_head)

if __name__ == "__main__":
    main()
```
