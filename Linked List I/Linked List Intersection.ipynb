#Linked List Intersection


Write a program which accepts two linked list containing integer data sorted in increasing order and returns a new linked list containing common elements from the two linked lists. Assume that there always be at least one element in common between the two linked lists.

Input and Output format:
The first line of the input is an integer 'n1' corresponds to the number of elements in the first linked list.
The next n1 lines correspond to the elements of the first linked list.
The next line of the input is an integer 'n2' corresponds to the number of elements in the second linked list.The next n2 lines correspond to the elements of the second linked list.

The output should be the common elements from the two linked list.


Sample Input :
6
1
2
3
4
5
6
4
2
4
6
8

Sample Output :
2
4
6



```python
class ListNode:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = ListNode(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def print_list(self):
        current = self.head
        while current:
            print(current.data)
            current = current.next

def find_intersection(list1, list2):
    head1 = list1.head
    head2 = list2.head
    result = LinkedList()

    while head1 and head2:
        if head1.data < head2.data:
            head1 = head1.next
        elif head1.data > head2.data:
            head2 = head2.next
        else:
            result.append(head1.data)
            head1 = head1.next
            head2 = head2.next

    return result

def main():
    # Input the first linked list
    n1 = int(input())
    list1 = LinkedList()
    for _ in range(n1):
        data = int(input())
        list1.append(data)

    # Input the second linked list
    n2 = int(input())
    list2 = LinkedList()
    for _ in range(n2):
        data = int(input())
        list2.append(data)

    # Find the intersection of the two linked lists
    intersection_list = find_intersection(list1, list2)

    # Print the result
    intersection_list.print_list()

if __name__ == "__main__":
    main()
```
