#Merge Two Sorted Linked list

Write a program to Merge two sorted singly linked lists into one sorted list.

Input / Output  Format:
Refer sample input-output for format specification.

Note: All text in bold refers to input and rest corresponds to output.

Sample input and Output:
Enter the element to insert in list 1:
100
Do you want insert again yes/no:
no
Enter the element to insert in list 2:
101
Do you want insert again yes/no:
no
List 1:
100
List 2:
101
Merged List in Ascending order:
100 101

```python
# Node class to define each node of the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to merge two sorted linked lists
def mergeLists(head1, head2):
    # Dummy node as the starting point of the merged list
    dummy = Node(0)
    current = dummy

    # Pointers for both lists
    current1 = head1
    current2 = head2

    # Traverse both lists and merge
    while current1 and current2:
        if current1.data <= current2.data:
            current.next = current1
            current1 = current1.next
        else:
            current.next = current2
            current2 = current2.next
        current = current.next

    # Attach remaining nodes of list1 or list2 if any
    if current1:
        current.next = current1
    if current2:
        current.next = current2

    return dummy.next

# Function to print the linked list
def printList(head):
    current = head
    while current:
        print(current.data, end=" ")
        current = current.next
    print()

# Main function to run the program
def main():
    # Input for list 1
    print("Enter the element to insert in list 1:")
    head1 = None
    while True:
        num = int(input().strip())
        new_node = Node(num)
        if head1 is None:
            head1 = new_node
            last_node1 = head1
        else:
            last_node1.next = new_node
            last_node1 = new_node
        print("Do you want insert again yes/no:")
        choice = input().strip().lower()
        if choice != "yes":
            break

    # Input for list 2
    print("Enter the element to insert in list 2:")
    head2 = None
    while True:
        num = int(input().strip())
        new_node = Node(num)
        if head2 is None:
            head2 = new_node
            last_node2 = head2
        else:
            last_node2.next = new_node
            last_node2 = new_node
        print("Do you want insert again yes/no:")
        choice = input().strip().lower()
        if choice != "yes":
            break

    # Print List 1
    print("List 1:")
    printList(head1)

    # Print List 2
    print("List 2:")
    printList(head2)

    # Merge and print merged list
    merged_head = mergeLists(head1, head2)
    print("Merged List in Ascending order:")
    printList(merged_head)

if __name__ == "__main__":
    main()
```
